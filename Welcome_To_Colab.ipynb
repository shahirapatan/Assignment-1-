implementation of 10 combinations of Data Operators b/w two arrays.

```
a = [10, 20, 30]
b = [2, 5, 3]

# 1 addition
add = [a[i] + b[i] for i in range(len(a))]

# 2 subtraction
sub = [a[i] - b[i] for i in range(len(a))]

# 3 multiplication
mul = [a[i] * b[i] for i in range(len(a))]

# 4 division
div = [a[i] / b[i] for i in range(len(a))]

# 5 floor division
floor_div = [a[i] // b[i] for i in range(len(a))]

# 6 modulo
mod = [a[i] % b[i] for i in range(len(a))]

# 7 mix + and *
mix1 = [(a[i] + b[i]) * 2 for i in range(len(a))]

# 8 mix - and %
mix2 = [(a[i] - b[i]) % 4 for i in range(len(a))]

# 9 mix * and //
mix3 = [(a[i] * 3) // b[i] for i in range(len(a))]

# 10 mix + and /
mix4 = [(a[i] + 10) / b[i] for i in range(len(a))]

print("Addition:", add)
print("Subtraction:", sub)
print("Multiplication:", mul)
print("Division:", div)
print("Floor Division:", floor_div)
print("Modulo:", mod)
print("Mix1:", mix1)
print("Mix2:", mix2)
print("Mix3:", mix3)
print("Mix4:", mix4)
```
